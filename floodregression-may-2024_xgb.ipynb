```python
# No missing values
# Data Manipulation: Bin all values > 10 into 10 bin.

#train = pd.read_csv('/kaggle/input/playground-series-s4e5/train.csv')
exclude_cols = ['FloodProbability', 'id']
label_col = 'FloodProbability'
for col in train.columns:
    if col not in exclude_cols:
        train[col] = train[col].apply(lambda x : 10 if x >= 10 else x)
        df = train.groupby([col]).agg({'FloodProbability': ['mean', 'max', 'min'], 'id': 'count'}).round(2).reset_index()
        df[col + 'dist'] = np.round(df['id'] / df['id'].sum(axis = 0) * 100)
        df.drop(columns= [col, 'id'], inplace = True)
        print(df)
```

# Estabilish a Baseline model

## Columns with high corelation

```python
import random
random.seed(47)
df = train
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(df.corr())
# We would need to start from all the columns intially
```

RandomForests Regressor on a sample population to see if we can overfit

```python
df.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import r2_score

from sklearn.model_selection import train_test_split
from sklearn.ensemble import HistGradientBoostingRegressor
import time
import xgboost as xgb
import matplotlib.pyplot as plt


rf = RandomForestRegressor()
gb = HistGradientBoostingRegressor()
#xgb = xgb.XGBRegressor()

X = df[[col for col in df.columns if col not in exclude_cols]]
Y = df[label_col]

def transform_columns(data):
    for col in data.columns: 
        if col not in exclude_cols:
            data[col] = data[col].apply(lambda x : 10 if x >= 10 else x)
    return data

def generate_submissions(model, num):
    X_submission = test[[col for col in test.columns if col not in exclude_cols]]
    y_submission = model.predict(X_submission)
    predictions = pd.DataFrame(data = y_submission, columns = ['FloodProbability'])
    submission = pd.concat([test, predictions], axis = 1)
    submission = submission[['id', 'FloodProbability']]
    submission.to_csv('/kaggle/working/model' + num + '.csv', index=False)
```

Experients on a sample of data

```python
start = time.time()
basemodel = rf.fit(X, Y)
end_time = time.time()
preds = rf.predict(X)
print('mae', np.round(mean_absolute_error(Y.values, np.array(preds)), 2))
print('Training time: ---', end_time - start)
```

```python
for col, imp in zip(X.columns, basemodel.feature_importances_):
    print(col, np.round(imp, 2))
```

```python
start = time.time()
gbmodel = gb.fit(X, Y)
end_time = time.time()
preds = gb.predict(X)
print(X.shape, Y.shape)
print('mape', np.round(mean_absolute_percentage_error(Y.values, np.array(preds)), 2))
print('Training time: ---', np.round(end_time - start))
```

Now, move on to baseline models on the whole data as verfitting was possible on a small subset

```python
X_t, X_test, y_t, y_test = train_test_split(X, Y, test_size = 0.2)
x_train, x_valid, y_train, y_valid = train_test_split(X_t, y_t, test_size = 0.125)
```

Moving to DL based techniques as the data is computationally large for scikit - learn based techniques

```python
gb = HistGradientBoostingRegressor(loss = 'squared_error', l2_regularization = 5, verbose=True, 
                                   categorical_features=[1] * len(X_t.columns), 
                                   random_state = 47, max_iter = 400, learning_rate = 0.5).fit(X_t, y_t)
```

```python
#print ('validation error:--', mean_absolute_error(y_valid, gb.predict(x_valid)))
predictions = gb.predict(X_test)
print('train set error: ', abs(gb.train_score_[-1]))
print('validation set error: ', abs(gb.validation_score_[-1]))
print ('test set error: ', np.sum((y_test - predictions)**2) / len(y_test))
print('Test set score: ',gb.score(X_test, y_test))
```

Xgboost Regressor

```python
for col in X_t.columns:
    if col not in exclude_cols:
        X_t[col] = X_t[col].astype('category') 

xb = xgb.XGBRegressor(enable_categorical = True, verbose = True, reg_alpha = 5, 
                      random_state = 47).fit(x_train, y_train)
```

```python
def squared_error(y_true, y_pred):    
    return np.sum((y_true - y_pred)**2)/ len(y_true)

train_preds = xb.predict(x_train)
valid_preds = xb.predict(x_valid)
test_preds = xb.predict(X_test)

train_score = r2_score(y_train, train_preds)
valid_score = r2_score(y_valid, valid_preds)
test_score = r2_score(y_test, test_preds)

train_error = squared_error(y_train, train_preds)
valid_error  = squared_error(y_valid, valid_preds)
test_error = squared_error(y_test, test_preds)

print('R2 Score:')
print(train_score, valid_score, test_score)
print("Squared Error:")
print(train_error, valid_error, test_error)
```
